```python
import seaborn as sns
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.displot(x="total_bill",kind="kde",data=df)
```

```python
sns.jointplot(x="total_bill",y="tip",data=df,hue="smoker")
```

```python
sns.catplot(x="day",data=df,y="tip",hue="smoker")
```

```python
sns.pairplot(data=df,palette="seismic",hue="smoker")
```

```python
sns.boxplot(data=df,x="day",y="total_bill",palette="viridis")
```

```python
sns.histplot(data=df,x="tip",bins=20)
```
